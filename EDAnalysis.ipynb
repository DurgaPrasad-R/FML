```python
#For Better Visualization
it_data = it_data.replace(['javascript','Javascript', 'js', 'JS'], 'JavaScript')
it_data = it_data.replace(['python', 'Python '], 'Python')
it_data = it_data.replace('c++', 'C++')
it_data = it_data.replace(['.net', '.Net'], '.NET')
it_data = it_data.replace(['Sql', 'sql'], 'SQL')
it_data = it_data.replace(['php','Php'], 'PHP')
it_data = it_data.replace('java', 'Java')
it_data = it_data.replace(['typescript', 'Typescript'], 'TypeScript')
```
